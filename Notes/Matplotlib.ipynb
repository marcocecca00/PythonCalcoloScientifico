```python
rng = np.random.default_rng(42)
```

Usiamo ora un codice equivalente per generare un insieme di coordinate $x$ ed $y$ che saranno uniti a schermo mediante una spezzata.
Con questa notazione separiamo i due oggetti, **figure** e **axes**:
- Una `Figur`e rappresenta l’intera figura
mostrata da Matplotlib
- Gli `Axes` sono i plot veri e propri (la spezzata per intenderci)

All’intero di ogni *Axes* ci sono due o tre oggetti di tipo *Axis*, che rappresentano gli assi cartesiani.


```python
def plot_x_y():    
    x = np.arange(1, 6)
    y = rng.integers(low=0, high=10, size=5)
    fig, ax = plt.subplots()
    ax.plot(x, y)
    return fig, ax

fig, ax = plot_x_y()
```

Possiamo dare un nome agli assi ed alla figura, impostando una griglia di riferimento.

```python
fig, ax = plot_x_y()
ax.set_title('Il nostro primo plot')        # Imposto il titolo
ax.set_xlabel('Asse delle X')               # Imposto il label per l'asse X
ax.set_ylabel('Asse delle Y')               # Imposto il label per l'asse Y
ax.grid(True)                               # Imposto la griglia di riferimento
plt.show()
```

**Esempio 1: plot di più funzioni**

Proviamo a plottare a schermo due funzioni, ovvero una retta ed una funzione sinusoidale.

```python
x = np.arange(0., 10., 0.01)
y_1 = 1 + 2 * x
y_2 = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y_1, label='Retta')
ax.plot(x, y_2, label='Funzione sinusoidale')

ax.set_title('Plot di due funzioni matematiche')
ax.set_xlabel('Asse x')
ax.set_ylabel('Asse y')

ax.legend()
ax.grid()

plt.show()
```

**Esempio 2: subplots**

Proviamo ad organizzare le funzioni che abbiamo plottato in precedenza in due diversi subplot.

```python
fig, (ax_1, ax_2) = plt.subplots(2, 1)
fig.suptitle('Due subplot di più funzioni matematiche')

# Primo subplot
ax_1.plot(x, y_1, label='Retta')
ax_1.set_ylabel('Asse y')
ax_1.legend()
ax_1.grid()

# Secondo subplot
ax_2.plot(x, y_2, label='Funzione sinusoidale')
ax_2.set_xlabel('Asse x')
ax_2.set_ylabel('Asse y')
ax_2.legend()
ax_2.grid()

# Mostro la figura
plt.show()
```

**Esempio 3: istogrammi**

Proviamo a visualizzare a schermo un istogramma.

```python
x = rng.integers(low=0, high=100, size=1000)
fig, ax = plt.subplots()
ax.hist(x, edgecolor='black', linewidth=1.2, density=True)

ax.set_xlabel('Bin')
ax.set_ylabel('Conteggio dei singoli elementi')
ax.set_title('Esempio di istogramma')

# Questa funzione ci permette di salvare la figura.
# plt.savefig('hist.png')

plt.show()
```
